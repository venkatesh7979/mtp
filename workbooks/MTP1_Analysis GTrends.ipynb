# Import Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Load the Data

```python
df = pd.read_csv(r'C:\Users\User\Desktop\MTP\Tweets and GT\GT\gtrends.csv')
```

```python
df['Day'] = pd.to_datetime(df['Day'])
```

```python
df.set_index('Day', inplace=True)
```

```python
import yfinance as yf
df_fd = yf.download("INR=X", start="2009-01-01", end="2022-11-11")
```

```python
df['INR=X_Close'] = df_fd['Close']
```

# Heatmap

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),annot=True)
```

```python
plt.figure()
plt.rcParams.update({'font.size': 10})
sns.heatmap(df[['inflation','interest rate','loan','jobs','price','save','exchange rate','INR','INR=X_Close']].corr(),annot=True,cmap='viridis')
plt.savefig(r'C:\Users\User\Desktop\MTP\Work Books\Images\corrgtrends.png',bbox_inches='tight',dpi=300)
```

```python
otherkeywords = ['CPI','vacation','shopping','cheap','price','ATM','cash','save','layoff']

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 15})

fig, axes = plt.subplots(nrows=len(otherkeywords), ncols=1)

for i in range(len(otherkeywords)):
    df[otherkeywords[i]].rolling(7).mean().plot(ax=axes[i],sharex=True,figsize=(15,23),ylabel=otherkeywords[i])
```

```python
fig.savefig(r'C:\Users\User\Desktop\MTP\Work Books\Images\gtrends2.png',bbox_inches='tight',dpi=300)
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 15})

fig, axes = plt.subplots(nrows=10, ncols=1)

df['inflation'].rolling(7).mean().plot(ax=axes[0],sharex=True,figsize=(15,23),ylabel='inflation')
df['GDP'].rolling(7).mean().plot(ax=axes[1],ylabel='GDP')
df['loan'].rolling(7).mean().plot(ax=axes[2],ylabel='loan')
df['jobs'].rolling(7).mean().plot(ax=axes[3],ylabel='jobs')
df['interest rate'].rolling(7).mean().plot(ax=axes[4],ylabel='interest rate')
df['exchange rate'].rolling(7).mean().plot(ax=axes[5],ylabel='exchange rate')
df['INRUSD'].rolling(7).mean().plot(ax=axes[6],ylabel='INRUSD')
df['USDINR'].rolling(7).mean().plot(ax=axes[7],ylabel='USDINR')
df['USD to INR'].rolling(7).mean().plot(ax=axes[8],ylabel='USD to INR')
df['INR=X_Close'].plot(ax=axes[9],ylabel='INR=X_Close')
```
